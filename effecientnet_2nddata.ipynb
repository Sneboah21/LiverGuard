```python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  1. DEPENDENCIES                                                        â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# If TensorFlow 2.x is not in the environment (e.g. Kaggle CPU notebooks) uncomment:
# %pip install -q tensorflow==2.15.0 tensorflow-addons seaborn

import os, random, warnings, numpy as np, matplotlib.pyplot as plt, seaborn as sns
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

warnings.filterwarnings('ignore')        # cleaner output
plt.style.use('seaborn-v0_8')

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  2. âš™ï¸  CONFIG                                                           â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
BASE_DIR = '/kaggle/input/liver-file/predicted_over22'   # â† change to your dataset root
USE_GPU  = True                                     # False â†’ comment-out mixed precision

IMG_SIZE   = (256, 256)
BATCH_SIZE = 16
SEED       = 42
AUTOTUNE   = tf.data.AUTOTUNE

train_dir = os.path.join(BASE_DIR, 'Train2')
val_dir   = os.path.join(BASE_DIR, 'Val2')
test_dir  = os.path.join(BASE_DIR, 'Test2')

# Reproducibility + optional mixed precision
tf.random.set_seed(SEED); np.random.seed(SEED); random.seed(SEED)
if USE_GPU:
    try:
        tf.keras.mixed_precision.set_global_policy('mixed_float16')
        print('âœ… Mixed precision enabled')
    except Exception as e:
        print('âš ï¸  Mixed precision not available:', e)

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  3. QUICK EDA                                                            â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
def plot_class_distribution(path, split_name):
    class_counts = {cls: len(os.listdir(os.path.join(path, cls)))
                    for cls in os.listdir(path)
                    if os.path.isdir(os.path.join(path, cls))}
    plt.figure(figsize=(4,3))
    plt.bar(class_counts.keys(), class_counts.values(), color=['steelblue','tomato'])
    plt.title(f'{split_name} class balance'); plt.ylabel('# images')
    plt.show()

for p, n in [(train_dir,'Train'), (val_dir,'Validation'), (test_dir,'Test')]:
    plot_class_distribution(p, n)     # based on snippet[1]

def plot_sample_grid(directory, n=4):
    classes = os.listdir(directory)
    plt.figure(figsize=(4*n, 4*len(classes)))
    for i, cls in enumerate(classes):
        imgs = random.sample(os.listdir(os.path.join(directory, cls)), n)
        for j, name in enumerate(imgs):
            img = tf.keras.utils.load_img(os.path.join(directory, cls, name),
                                          target_size=IMG_SIZE)
            ax = plt.subplot(len(classes), n, i*n + j + 1)
            plt.imshow(img); plt.axis(False)
            if j==0: ax.set_ylabel(cls, fontsize=14)
    plt.suptitle('Random training samples', fontsize=18); plt.tight_layout(); plt.show()

plot_sample_grid(train_dir)           # based on snippet[1]

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  4. tf.data INPUT PIPELINE                                               â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
def build_dataset(path, training=False):
    ds = tf.keras.utils.image_dataset_from_directory(
            path, image_size=IMG_SIZE, batch_size=BATCH_SIZE,
            label_mode='binary', shuffle=training, seed=SEED)
    preprocess = tf.keras.applications.xception.preprocess_input
    augment    = tf.keras.Sequential([
                    tf.keras.layers.RandomFlip('horizontal'),
                    tf.keras.layers.RandomRotation(0.05),
                    tf.keras.layers.RandomZoom(0.10)
                 ])
    ds = ds.map(lambda x,y: (preprocess(x), y), num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.map(lambda x,y: (augment(x, training=True), y),
                    num_parallel_calls=AUTOTUNE)
    return ds.prefetch(AUTOTUNE)

train_ds = build_dataset(train_dir, training=True)
val_ds   = build_dataset(val_dir)
test_ds  = build_dataset(test_dir)

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  5. MODEL: Xception + custom head                                        â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
base = tf.keras.applications.Xception(weights='imagenet',
                                      include_top=False,
                                      input_shape=IMG_SIZE+(3,))
base.trainable = False                             # Phase-1: frozen

inputs  = tf.keras.Input(shape=IMG_SIZE+(3,))
x = base(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.3)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid',
                                dtype='float32')(x)  # keep float32 for stability

model = tf.keras.Model(inputs, outputs)
model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  6. TRAINING                                                             â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint('xception_best.h5', save_best_only=True),
    tf.keras.callbacks.ReduceLROnPlateau(patience=3, factor=0.2, verbose=1)
]

print('\nðŸŸ¢  Phase-1: train head (base frozen)')
hist1 = model.fit(train_ds, validation_data=val_ds,
                  epochs=30, callbacks=callbacks)

# Phase-2: fine-tune last 30 layers
base.trainable = True
for layer in base.layers[:-30]:
    layer.trainable = False

model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])
print('\nðŸŸ¢  Phase-2: fine-tune last 30 layers')
hist2 = model.fit(train_ds, validation_data=val_ds,
                  epochs=30, callbacks=callbacks)

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  7. LEARNING CURVES                                                      â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
def plot_history(h1, h2):
    fig, ax = plt.subplots(1,2, figsize=(12,4))
    for h, lbl in zip([h1,h2], ['Frozen','Fine-tuned']):
        ep = range(1, len(h.history['accuracy'])+1)
        ax[0].plot(ep, h.history['accuracy'], label=f'{lbl}-train')
        ax[0].plot(ep, h.history['val_accuracy'], '--', label=f'{lbl}-val')
        ax[1].plot(ep, h.history['loss'],     label=f'{lbl}-train')
        ax[1].plot(ep, h.history['val_loss'], '--', label=f'{lbl}-val')
    ax[0].set_title('Accuracy'); ax[1].set_title('Loss')
    for a in ax: a.grid(True); a.legend()
    plt.show()

plot_history(hist1, hist2)

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  8. TEST-SET EVALUATION                                                  â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
test_loss, test_acc = model.evaluate(test_ds, verbose=0)
print(f'\nâœ…  Test Accuracy: {test_acc:.3%} | Loss: {test_loss:.4f}')

y_prob = model.predict(test_ds).ravel()
y_pred = (y_prob > 0.5).astype(int)
y_true = np.concatenate([y for _, y in test_ds], axis=0)

print('\n', classification_report(y_true, y_pred,
                                  target_names=['Non tumour', 'Tumour']))

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non tumour','Tumour'],
            yticklabels=['Non tumour','Tumour'])
plt.title('Confusion Matrix'); plt.ylabel('True'); plt.xlabel('Predicted'); plt.show()

fpr, tpr, _ = roc_curve(y_true, y_prob)
plt.figure(); plt.plot(fpr, tpr, label=f'AUC = {auc(fpr,tpr):.3f}')
plt.plot([0,1],[0,1],'k--'); plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate'); plt.title('ROC Curve'); plt.legend(); plt.show()

# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  9. SAVE BEST MODEL                                                      â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
model.save('xception_liver_seg.h5')
print('ðŸ’¾  Model saved to xception_liver_seg.h5')
```
