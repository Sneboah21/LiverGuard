```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import shutil
from pathlib import Path

# ====================================================================
# STEP 1: Copy dataset to working directory (writable in Kaggle)
# ====================================================================

# Kaggle input directory (read-only)
input_base = '/kaggle/input/brain-tumor-mri-dataset'

# Kaggle working directory (writable)
working_base = '/kaggle/working/brain-tumor-dataset'

# Copy training and testing folders to working directory
if not os.path.exists(working_base):
    print("Copying dataset to working directory...")
    shutil.copytree(os.path.join(input_base, 'Training'), 
                    os.path.join(working_base, 'Training'))
    shutil.copytree(os.path.join(input_base, 'Testing'), 
                    os.path.join(working_base, 'Testing'))
    print("Dataset copied successfully!")
else:
    print("Dataset already exists in working directory")

# ====================================================================
# STEP 2: Create validation split from training data
# ====================================================================

def create_validation_split(train_dir, val_dir, val_split=0.2, seed=42):
    """
    Splits training data into train and validation sets
    """
    if os.path.exists(val_dir):
        print(f"Validation directory already exists: {val_dir}")
        return
    
    os.makedirs(val_dir, exist_ok=True)
    
    # Get all class folders
    classes = [d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))]
    
    for class_name in classes:
        class_train_path = os.path.join(train_dir, class_name)
        class_val_path = os.path.join(val_dir, class_name)
        os.makedirs(class_val_path, exist_ok=True)
        
        # Get all images in this class
        images = [f for f in os.listdir(class_train_path) 
                 if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        # Split images
        train_imgs, val_imgs = train_test_split(images, test_size=val_split, random_state=seed)
        
        # Move validation images to validation folder
        for img in val_imgs:
            src = os.path.join(class_train_path, img)
            dst = os.path.join(class_val_path, img)
            shutil.move(src, dst)
        
        print(f"Class '{class_name}': {len(train_imgs)} training, {len(val_imgs)} validation")

# Set directory paths
train_dir = os.path.join(working_base, 'Training')
val_dir = os.path.join(working_base, 'Validation')
test_dir = os.path.join(working_base, 'Testing')

# Create validation split
create_validation_split(train_dir, val_dir, val_split=0.2)

# ====================================================================
# STEP 3: Data Generators with Correct Augmentation
# ====================================================================

# Training data: Augmentation + Rescaling
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Validation and Test data: Only Rescaling (NO augmentation)
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Create generators
train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)

val_gen = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)

test_gen = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

print(f"\nTraining samples: {train_gen.samples}")
print(f"Validation samples: {val_gen.samples}")
print(f"Test samples: {test_gen.samples}")
print(f"Classes: {list(train_gen.class_indices.keys())}\n")

# ====================================================================
# STEP 4: Build Model with Pretrained VGG16
# ====================================================================

# Hyperparameters
dropout_rate = 0.5
dense_units = 256
learning_rate = 1e-4

# Load pretrained VGG16
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

# Build model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(dense_units, activation='relu', kernel_regularizer='l2'),
    Dropout(dropout_rate),
    Dense(4, activation='softmax')
])

model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=learning_rate),
    metrics=['accuracy']
)

print("Model Architecture:")
model.summary()

# ====================================================================
# STEP 5: Training with Callbacks
# ====================================================================

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)
lr_reduce = ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.3, verbose=1)
checkpoint = ModelCheckpoint('best_vgg16_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)

# Phase 1: Train with frozen base
print("\n" + "="*70)
print("PHASE 1: Training with frozen VGG16 base")
print("="*70 + "\n")

history = model.fit(
    train_gen,
    epochs=25,
    validation_data=val_gen,
    callbacks=[early_stop, lr_reduce, checkpoint],
    verbose=1
)

# Phase 2: Fine-tuning
print("\n" + "="*70)
print("PHASE 2: Fine-tuning VGG16")
print("="*70 + "\n")

base_model.trainable = True
for layer in base_model.layers[:-4]:
    layer.trainable = False

model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(learning_rate=1e-5),
    metrics=['accuracy']
)

history_finetune = model.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen,
    callbacks=[early_stop, lr_reduce],
    verbose=1
)

# ====================================================================
# STEP 6: Evaluation on Test Set
# ====================================================================

print("\n" + "="*70)
print("EVALUATING ON TEST SET")
print("="*70 + "\n")

test_loss, test_acc = model.evaluate(test_gen, verbose=1)
print(f"\n{'='*70}")
print(f"Final Test Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")
print(f"Final Test Loss: {test_loss:.4f}")
print(f"{'='*70}\n")

# Get predictions
predictions = model.predict(test_gen, verbose=1)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_gen.classes
class_labels = list(test_gen.class_indices.keys())

# Classification Report
print("\n" + "="*70)
print("CLASSIFICATION REPORT")
print("="*70 + "\n")
print(classification_report(true_classes, predicted_classes, target_names=class_labels))

# Confusion Matrix
cm = confusion_matrix(true_classes, predicted_classes)
print("\n" + "="*70)
print("CONFUSION MATRIX")
print("="*70 + "\n")
print(cm)

# ====================================================================
# STEP 7: Visualizations
# ====================================================================

# Combine history
acc = history.history['accuracy'] + history_finetune.history['accuracy']
val_acc = history.history['val_accuracy'] + history_finetune.history['val_accuracy']
loss = history.history['loss'] + history_finetune.history['loss']
val_loss = history.history['val_loss'] + history_finetune.history['val_loss']
epochs_range = range(len(acc))

# Accuracy and Loss plots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].plot(epochs_range, acc, label='Training Accuracy', linewidth=2, color='#2E86AB')
axes[0].plot(epochs_range, val_acc, label='Validation Accuracy', linewidth=2, color='#A23B72')
axes[0].axvline(x=len(history.history['accuracy']), color='red', linestyle='--', 
                linewidth=1.5, label='Fine-tuning starts', alpha=0.7)
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Accuracy', fontsize=12)
axes[0].set_title('Training & Validation Accuracy', fontsize=14, fontweight='bold')
axes[0].legend(loc='lower right')
axes[0].grid(True, alpha=0.3)

axes[1].plot(epochs_range, loss, label='Training Loss', linewidth=2, color='#2E86AB')
axes[1].plot(epochs_range, val_loss, label='Validation Loss', linewidth=2, color='#A23B72')
axes[1].axvline(x=len(history.history['loss']), color='red', linestyle='--', 
                linewidth=1.5, label='Fine-tuning starts', alpha=0.7)
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Loss', fontsize=12)
axes[1].set_title('Training & Validation Loss', fontsize=14, fontweight='bold')
axes[1].legend(loc='upper right')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
plt.show()

# Confusion Matrix Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_labels, yticklabels=class_labels,
            cbar_kws={'label': 'Count'}, linewidths=0.5, linecolor='gray')
plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
plt.ylabel('True Label', fontsize=12, fontweight='bold')
plt.title(f'Confusion Matrix on Test Set\nAccuracy: {test_acc:.4f}', 
          fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# ====================================================================
# STEP 8: Save Final Model
# ====================================================================

model.save('final_brain_tumor_vgg16_model.h5')
print("\nFinal model saved as 'final_brain_tumor_vgg16_model.h5'")
```
